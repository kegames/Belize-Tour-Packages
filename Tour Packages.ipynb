```python
from bs4 import BeautifulSoup
import requests
import csv

source = requests.get('https://www.viator.com/San-Ignacio-attractions/Blue-Hole-National-Park/d5083-a16566').text

soup = BeautifulSoup(source, 'lxml')

csv_file = open('tour_scrape.csv', 'w')
csv_writer = csv.writer(csv_file)
csv_writer.writerow(['Title', 'Rating', 'Price', 'Description', 'Time Complete'])

#print(soup.prettify())

#for header in soup.find_all('div', class_ = 'tracked-elements'):
for header in soup.find_all('div', class_ = 'product-card-inner-row row m-0'):
    
    try:
        title = header.find('h2', class_= 'product-card-row-title mb-0 pt-md-4').text
    except Exception as e:
        title = None
        
    #print(title)
    
    try:
        rating = header.find('span', class_='small text-body text-nowrap reviewLink').text
    except Exception as e:
        rating = None
    
    #print(rating)
    
    try:
        price = header.find('div', class_='h3 line-height-same mb-0 price-font text-md-right').text
    except Exception as e:
        price = None
        
    #print(price)
    
    #description = header.find('div', class_='d-none d-md-flex flex-column mt-3 mt-md-2 prod-desc-container').div.text
    try:
        description = header.find('div', class_='text-body summary-text mb-0').text
    except Exception as e:
        description = None
        
    #print(description)

    try:
        timeComplete = header.find('span', class_="px-1 px-1 align-middle product-card-footer-text").span.text
    except Exception as e:
        timeComplete = None
        
    #print(timeComplete)
    
    
    csv_writer.writerow([title, rating, price, description, timeComplete])

csv_file.close()
    
    
```

```python
#Bar Chart for Prices based on Category
import csv
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from matplotlib.ticker import StrMethodFormatter

def add_value_labels(ax, spacing=5):

    for rect in ax.patches:
        y_value = rect.get_height()
        x_value = rect.get_x() + rect.get_width()/2
        
        space = spacing
        va = 'bottom'
       
        label = "${:.02f}".format(y_value)
        
        ax.annotate(
            label,
            (x_value, y_value),
            xytext=(0, space),
            textcoords = "offset points",
            ha='center',
            va=va
        )
        
df = pd.read_csv('tour_scrape_master_v2.csv')
df = df.dropna()

df["Price of Tour"] = [x.strip('$') for x in df["Price"]]
df["Price of Tour"] = pd.to_numeric(df["Price of Tour"], errors='coerce')


x = df.groupby('Category')['Price of Tour'].mean().sort_values()

print(x) 

ax = x.plot(kind='bar', figsize=(13,9), color='#86bf91', zorder=2, width=0.65)

# Despine
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['bottom'].set_visible(False)

# Switch off ticks
ax.tick_params(axis="both", which="both", bottom="off", top="off", labelbottom="on", left="off", right="off", labelleft="on")


# Set x-axis label
ax.set_xlabel("Average Price of Tour by Category", labelpad=20, weight='bold', size=15)

# Set y-axis label
ax.set_ylabel("Price", labelpad=20, weight='bold', size=15)

# Format y-axis label
ax.yaxis.set_major_formatter('${x:1.2f}')  # Format the price.

add_value_labels(ax)
  

```
